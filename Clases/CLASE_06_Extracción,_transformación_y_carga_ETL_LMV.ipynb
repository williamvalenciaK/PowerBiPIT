>* La cadena de texto que se va a truncar y el nmero de caracteres que se desean extraer.

#### **<font color="#FFFFFF">Operadores de comparacin en Lenguaje M</font>**

</td>
</tr>
</table>

| Operador | Nombre del Operador | Descripcin del Operador de Comparacin en Lenguaje M de Power BI |
|----------|---------------------|-------------------------------------------------------------------|
| =       | Igual               | Devuelve verdadero si los dos valores son iguales                   |
| <>       | Diferente           | Devuelve verdadero si los dos valores son diferentes                |
| <        | Menor que           | Devuelve verdadero si el valor de la izquierda es menor que el de la derecha |
| <=       | Menor o igual que   | Devuelve verdadero si el valor de la izquierda es menor o igual que el de la derecha |
| >        | Mayor que           | Devuelve verdadero si el valor de la izquierda es mayor que el de la derecha |
| >=       | Mayor o igual que   | Devuelve verdadero si el valor de la izquierda es mayor o igual que el de la derecha |
| &&       | Y                   | Devuelve verdadero si ambas condiciones son verdaderas              |
| \|\|       | O                   | Devuelve verdadero si al menos una de las condiciones es verdadera    |



```
Date.EndOfMonth
(
    Date.FromText
    (
        "01/"
        &
        Text.From
        (
                 if Text.Start([Atributo],3)="Ene" then "01"
            else if Text.Start([Atributo],3)="Feb" then "02"
            else if Text.Start([Atributo],3)="Mar" then "03"
            else if Text.Start([Atributo],3)="Abr" then "04"
            else if Text.Start([Atributo],3)="May" then "05"
            else if Text.Start([Atributo],3)="Jun" then "06"
            else if Text.Start([Atributo],3)="Jul" then "07"
            else if Text.Start([Atributo],3)="Ago" then "08"
            else if Text.Start([Atributo],3)="Set" then "09"
            else if Text.Start([Atributo],3)="Oct" then "10"
            else if Text.Start([Atributo],3)="Nov" then "11"
            else if Text.Start([Atributo],3)="Dic" then "12"
            else ""
        )
        &
        "/20"
        &
        Text.End([Atributo],2)   // 21
    )
)
```
