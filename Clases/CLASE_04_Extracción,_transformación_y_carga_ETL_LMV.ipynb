La combinación de consultas es útil cuando tienes varias fuentes de datos relacionadas que deseas combinar en una sola tabla para su análisis.

En Power Query, puedes combinar consultas utilizando diferentes tipos de combinaciones, como la combinación por columnas coincidentes o por relaciones establecidas previamente.

Puedes especificar las columnas clave que se utilizarán para realizar la combinación y elegir el tipo de combinación que deseas realizar, como una combinación interna, externa izquierda, externa derecha, anticombinación izquierda, anticombinación derecha.

Una vez que hayas combinado las consultas, puedes realizar transformaciones adicionales en la tabla resultante, como agregar columnas calculadas, filtrar filas o cambiar los tipos de datos.

Luego, puedes cargar la tabla combinada en el modelo de datos de Power BI para su análisis y visualización.

La combinación externa izquierda, también conocida como "Left Outer Join" en inglés, es un tipo de combinación de consultas en Power BI.

Es una operación que permite unir dos tablas o consultas basándose en una columna común, pero además incluye todos los registros de la tabla de la izquierda, incluso si no hay coincidencias en la tabla de la derecha.

## **<font color="#000000"> ◉ Combinación externa derecha</font>**

</td>
</tr>
</table>

La combinación externa derecha, también conocida como "Right Outer Join" en inglés, es otro tipo de combinación de consultas en Power BI.

Es una operación que permite unir dos tablas o consultas basándose en una columna común, pero además incluye todos los registros de la tabla de la derecha, incluso si no hay coincidencias en la tabla de la izquierda.

En una combinación externa derecha, los registros de la tabla de la derecha se mantienen en el resultado final, sin importar si hay coincidencias en la tabla de la izquierda.

Si no hay coincidencias, se rellenarán los valores correspondientes con valores nulos o en blanco en las columnas de la tabla de la izquierda.
