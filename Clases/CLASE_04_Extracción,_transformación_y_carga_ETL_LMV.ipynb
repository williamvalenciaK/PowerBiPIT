La combinacin de consultas es til cuando tienes varias fuentes de datos relacionadas que deseas combinar en una sola tabla para su anlisis.

En Power Query, puedes combinar consultas utilizando diferentes tipos de combinaciones, como la combinacin por columnas coincidentes o por relaciones establecidas previamente.

Puedes especificar las columnas clave que se utilizarn para realizar la combinacin y elegir el tipo de combinacin que deseas realizar, como una combinacin interna, externa izquierda, externa derecha, anticombinacin izquierda, anticombinacin derecha.

Una vez que hayas combinado las consultas, puedes realizar transformaciones adicionales en la tabla resultante, como agregar columnas calculadas, filtrar filas o cambiar los tipos de datos.

Luego, puedes cargar la tabla combinada en el modelo de datos de Power BI para su anlisis y visualizacin.

## **<font color="#000000">  Combinacin externa derecha</font>**

</td>
</tr>
</table>

La combinacin externa derecha, tambin conocida como "Right Outer Join" en ingls, es otro tipo de combinacin de consultas en Power BI.

Es una operacin que permite unir dos tablas o consultas basndose en una columna comn, pero adems incluye todos los registros de la tabla de la derecha, incluso si no hay coincidencias en la tabla de la izquierda.

En una combinacin externa derecha, los registros de la tabla de la derecha se mantienen en el resultado final, sin importar si hay coincidencias en la tabla de la izquierda.

Si no hay coincidencias, se rellenarn los valores correspondientes con valores nulos o en blanco en las columnas de la tabla de la izquierda.
