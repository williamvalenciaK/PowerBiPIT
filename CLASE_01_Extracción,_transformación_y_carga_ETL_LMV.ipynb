>* Business Intelligence (BI), o inteligencia de negocios, se refiere al conjunto de tecnologas, herramientas y prcticas utilizadas para recopilar, integrar, analizar y presentar datos empresariales con el fin de obtener informacin valiosa para la toma de decisiones estratgicas y operativas en una organizacin.

>* Para lograr esto, el BI utiliza tcnicas de extraccin, transformacin y carga de datos (ETL), almacenamiento y gestin de datos, anlisis de datos y visualizacin de informacin.

Los datos en el contexto del Business Intelligence (BI) son piezas de informacin recopiladas y almacenadas por una organizacin con el objetivo de utilizarlas para el anlisis y la toma de decisiones.

A continuacin, se describen las principales vistas disponibles en Power BI Desktop:

>* **Vista de informe (Report view)**: Esta vista es la principal en Power BI Desktop y se utiliza para disear y crear informes interactivos. Permite agregar visualizaciones, como grficos, tablas, mapas y otros elementos visuales, y configurar sus propiedades. En la vista de informe, se pueden aplicar filtros, crear relaciones entre diferentes conjuntos de datos y disear paneles de navegacin.

Permite realizar acciones como la limpieza de datos, la unin de tablas, la filtracin y la aplicacin de transformaciones avanzadas utilizando el Editor de consultas de Power Query.
