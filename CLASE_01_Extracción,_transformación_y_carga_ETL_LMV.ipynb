## **<font color="#000000">¿Qué es la Inteligencia de Negocios?</font>**

</td>
</tr>
</table>

<center>

imagen.png

</center>

>* Business Intelligence (BI), o inteligencia de negocios, se refiere al conjunto de tecnologías, herramientas y prácticas utilizadas para recopilar, integrar, analizar y presentar datos empresariales con el fin de obtener información valiosa para la toma de decisiones estratégicas y operativas en una organización.
>* El objetivo principal del Business Intelligence es convertir los datos en conocimientos accionables que ayuden a las empresas a mejorar su rendimiento, identificar oportunidades de crecimiento, optimizar sus operaciones y obtener ventajas competitivas en el mercado.
>* Para lograr esto, el BI utiliza técnicas de extracción, transformación y carga de datos (ETL), almacenamiento y gestión de datos, análisis de datos y visualización de información.
>* El proceso de Business Intelligence comienza con la recopilación de datos de diversas fuentes internas y externas de la organización, como bases de datos, sistemas transaccionales, redes sociales, fuentes en línea, entre otros.
>* Estos datos se transforman y se almacenan en un almacén de datos o data warehouse, donde se pueden combinar y relacionar para su análisis.
>* Una vez que los datos están disponibles en el almacén de datos, se aplican técnicas de análisis y minería de datos para descubrir patrones, tendencias y relaciones ocultas en los datos.
>* Estos análisis pueden incluir consultas ad hoc, informes estándar, análisis estadístico, minería de datos, análisis predictivo y otras técnicas avanzadas.
>* La visualización de datos es otra parte importante del Business Intelligence.
>* Los resultados del análisis se presentan en informes interactivos, paneles de control y visualizaciones gráficas para que los usuarios puedan comprender rápidamente la información y tomar decisiones informadas.

## **<font color="#000000">¿Qué son los datos?</font>**

</td>
</tr>
</table>

Los datos en el contexto del Business Intelligence (BI) son piezas de información recopiladas y almacenadas por una organización con el objetivo de utilizarlas para el análisis y la toma de decisiones.

Los datos pueden provenir de diversas fuentes, tanto internas como externas a la organización, y pueden ser estructurados, semiestructurados o no estructurados.

En el BI, los datos se consideran el combustible que impulsa el proceso de obtención de conocimientos y la generación de informes.

Estos datos pueden incluir información sobre ventas, clientes, inventarios, transacciones financieras, recursos humanos, marketing, operaciones y cualquier otro aspecto relevante del negocio.

Es importante destacar que los datos en el BI deben ser precisos, completos, consistentes y estar actualizados para garantizar la calidad de los análisis y las decisiones resultantes.

Además, **los datos deben estar bien organizados y estructurados** de manera que se puedan utilizar eficientemente en las herramientas y los sistemas de BI.

Algunos ejemplos de datos utilizados en el BI incluyen:

>* **Datos transaccionales**: Información detallada sobre las transacciones comerciales, como ventas, compras, pagos y facturación.
>* **Datos de clientes**: Datos demográficos, historiales de compras, preferencias, comportamiento en línea y otra información relacionada con los clientes.
>* **Datos operativos**: Información sobre las operaciones diarias de la organización, como producción, logística, inventario y gestión de la cadena de suministro.
>* **Datos financieros**: Información contable, estados financieros, presupuestos, costos y otros datos relacionados con la situación financiera de la empresa.
>* **Datos de marketing**: Datos de campañas de marketing, análisis de mercado, comportamiento del consumidor y métricas de rendimiento de marketing.
>* **Datos externos**: Información de fuentes externas, como datos demográficos, indicadores económicos, datos de la competencia y tendencias de la industria.

## **<font color="#000000">¿Qué es la información?</font>**

</td>
</tr>
</table>

La información se refiere al resultado del procesamiento y análisis de los datos recopilados por una organización.

La información se genera a partir de la interpretación y el contexto aplicado a los datos, lo que permite obtener conocimientos significativos y relevantes que respaldan la toma de decisiones empresariales.

La información en el BI se presenta de manera estructurada y significativa, lo que facilita la comprensión y el uso por parte de los usuarios y tomadores de decisiones.

La información proporciona respuestas a preguntas específicas y brinda una visión clara del rendimiento, las tendencias y los patrones del negocio.

Las características clave de la información en el BI incluyen:

>* **Contexto y relevancia**: La información se presenta en el contexto adecuado y es relevante para el área o el problema específico que se está abordando. Se filtran y resumen los datos relevantes para proporcionar información valiosa.
>* **Accesibilidad y disponibilidad**: La información en el BI está fácilmente disponible para los usuarios autorizados en la organización. Se accede a través de herramientas y sistemas de BI que permiten la consulta, el análisis y la generación de informes de manera eficiente.
>* **Claridad y comprensión**: La información se presenta de manera clara y comprensible, utilizando visualizaciones gráficas, informes estructurados y paneles de control interactivos. Esto facilita la interpretación y el análisis de la información por parte de los usuarios.
>* **Precisión y confiabilidad**: La información en el BI se basa en datos precisos y confiables. Se aplican técnicas de limpieza, validación y verificación de datos para garantizar la calidad y la integridad de la información generada.
>* **Accionabilidad**: La información en el BI se presenta de manera que pueda tomarse acción basada en ella. Proporciona ideas y recomendaciones claras que ayudan a los tomadores de decisiones a planificar estrategias, optimizar operaciones y alcanzar los objetivos empresariales.

La información agrega contexto y significado a los datos, proporcionando conocimientos útiles y orientación para la toma de decisiones en el BI.

## **<font color="#000000">¿Qué es conocimiento?</font>**

</td>
</tr>
</table>

El conocimiento se refiere a la comprensión profunda y contextualizada de la información generada a través del análisis de datos.

El conocimiento es el nivel más alto de valor agregado en el proceso de BI, ya que va más allá de la simple presentación de información y proporciona una comprensión significativa de los patrones, las relaciones y las implicaciones comerciales.

El conocimiento en el BI implica la capacidad de interpretar la información y extraer ideas valiosas que pueden impulsar la toma de decisiones estratégicas y operativas en una organización.

A través del análisis y la comprensión de la información, se obtiene una visión más profunda de los problemas y las oportunidades empresariales.

Algunos aspectos clave del conocimiento en el BI incluyen:

>* **Interpretación**: El conocimiento implica la capacidad de interpretar la información en el contexto adecuado y comprender su significado en relación con los objetivos y desafíos empresariales.
>* **Contextualización**: El conocimiento se basa en la capacidad de relacionar la información con el entorno empresarial, los datos históricos, las tendencias de la industria y otros factores relevantes.
>* **Identificación de patrones y tendencias**: El conocimiento implica la capacidad de identificar patrones, tendencias y relaciones significativas en los datos, lo que permite entender mejor el comportamiento del negocio y las oportunidades que pueden surgir.
>* **Generación de ideas y recomendaciones**: El conocimiento en el BI se utiliza para generar nuevas ideas, estrategias y recomendaciones basadas en la comprensión profunda de la información y su aplicación al contexto empresarial.
>* **Aprendizaje continuo**: El conocimiento en el BI implica un enfoque de aprendizaje continuo, donde se utilizan los resultados y las lecciones aprendidas de las decisiones pasadas para mejorar y ajustar las estrategias futuras.

## **<font color="#000000">¿Qué son las decisiones?</font>**

</td>
</tr>
</table>

Las decisiones se refieren a las elecciones tomadas por los líderes y tomadores de decisiones de una organización basadas en el análisis y la comprensión de la información generada a través del BI.

Estas decisiones están orientadas a lograr los objetivos empresariales, mejorar el rendimiento y responder a los desafíos y oportunidades identificados a través del análisis de datos.

Las decisiones en el contexto del BI pueden abarcar diferentes áreas y niveles de la organización, desde decisiones estratégicas a largo plazo hasta decisiones operativas diarias.

Algunos ejemplos de decisiones que se toman utilizando el BI incluyen:

>* **Decisiones estratégicas**: Estas decisiones están relacionadas con los objetivos y la dirección general de la organización. Pueden incluir decisiones sobre la entrada a nuevos mercados, la adquisición de empresas, la diversificación de productos o la expansión geográfica. El BI proporciona información valiosa para evaluar diferentes escenarios y tomar decisiones informadas sobre la estrategia empresarial.
>* **Decisiones tácticas**: Estas decisiones están relacionadas con la implementación de la estrategia y la asignación de recursos. Pueden incluir decisiones sobre la asignación de presupuestos, la planificación de la fuerza laboral, la gestión de inventarios o la optimización de la cadena de suministro. El BI proporciona información detallada y actualizada que ayuda a tomar decisiones tácticas eficientes y efectivas.
>* **Decisiones operativas**: Estas decisiones se toman a nivel diario para gestionar las actividades operativas de la organización. Pueden incluir decisiones sobre la programación de producción, la gestión de ventas, la atención al cliente o la gestión del flujo de efectivo. El BI proporciona información en tiempo real que permite tomar decisiones operativas rápidas y basadas en datos.
>* **Decisiones de marketing**: Estas decisiones están relacionadas con las estrategias de marketing y promoción de la organización. Pueden incluir decisiones sobre segmentación de mercado, posicionamiento de productos, estrategias de precios o campañas publicitarias. El BI proporciona información sobre el comportamiento del consumidor, las preferencias y las tendencias del mercado, lo que ayuda a tomar decisiones de marketing más efectivas.

## **<font color="#000000">¿Qué es Power BI?</font>**

</td>
</tr>
</table>

**Power BI** es una plataforma de análisis empresarial desarrollada por Microsoft que ofrece una amplia gama de capacidades para visualizar y analizar datos.

A continuación, se presentan algunas de las principales cosas que Power BI puede hacer:

>* **Conexión a varias fuentes de datos**: Power BI puede conectarse a una variedad de fuentes de datos, como bases de datos relacionales, servicios en la nube, archivos de Excel, SharePoint y muchas otras. Puede extraer datos de estas fuentes y combinarlos para su análisis.
>* **Modelado de datos**: Power BI permite modelar los datos importados o conectados para crear relaciones entre ellos. Puede realizar transformaciones, aplicar reglas comerciales y crear medidas y columnas calculadas para enriquecer los datos.
>* **Creación de visualizaciones interactivas**: Power BI proporciona una amplia gama de visualizaciones, como gráficos, tablas, mapas, gráficos de dispersión y muchas más. Puede arrastrar y soltar los campos de datos en el lienzo para crear visualizaciones interactivas y personalizadas.
>* **Creación de informes y paneles**: Puede diseñar informes interactivos utilizando las visualizaciones creadas y organizarlos en paneles. Los paneles son tableros de instrumentos que le permiten ver múltiples informes y visualizaciones en un solo lugar.
>* **Compartir y colaborar**: Puede compartir informes y paneles con otros usuarios dentro y fuera de su organización. Puede controlar los permisos de acceso y permitir que las personas vean, interactúen o editen los informes según sus roles.
>* **Análisis ad-hoc**: Power BI permite realizar análisis ad-hoc explorando los datos de forma interactiva. Puede filtrar, ordenar y resumir los datos según sus necesidades y realizar análisis en tiempo real.
>* **Publicación y distribución**: Puede publicar informes y paneles en el servicio en la nube de Power BI para que estén disponibles en línea. También puede programar actualizaciones automáticas de datos y enviar informes por correo electrónico a los destinatarios.
>* **Integración con otras aplicaciones de Microsoft**: Power BI se integra estrechamente con otras aplicaciones de Microsoft, como Excel, SharePoint, Teams y Azure. Puede importar datos desde estas aplicaciones y compartir informes en ellas.

## **<font color="#000000">Tipos de licenciamiento</font>**

</td>
</tr>
</table>

Power BI ofrece diferentes opciones de licenciamiento para adaptarse a las necesidades y requisitos de diferentes usuarios y organizaciones.

A continuación se presentan los principales tipos de licenciamiento de Power BI:

>* **Power BI Desktop**: Power BI Desktop es una herramienta gratuita que permite a los usuarios crear informes y visualizaciones en sus computadoras locales. No se requiere licencia para usar Power BI Desktop, y los informes creados se pueden compartir y ver utilizando otras opciones de licenciamiento de Power BI.
>* **Power BI Pro**: Power BI Pro es una licencia de usuario individual que brinda acceso completo a las capacidades de creación, colaboración y consumo de Power BI. Con una licencia Pro, los usuarios pueden compartir informes y paneles con otros usuarios, colaborar en tiempo real y acceder a contenido compartido por otros. Esta licencia requiere una suscripción mensual o anual por usuario.
>* **Power BI Premium**: Power BI Premium está diseñado para organizaciones que requieren capacidades empresariales a escala. Con una licencia Premium, los usuarios tienen acceso completo a todas las características de Power BI Pro y también pueden compartir informes y paneles con usuarios gratuitos. Además, Power BI Premium ofrece un rendimiento mejorado, mayor capacidad de almacenamiento y opciones de administración avanzadas. El licenciamiento de Power BI Premium se basa en la capacidad y se factura mensualmente.
>* **Power BI Embedded**: Power BI Embedded permite a los desarrolladores integrar informes y paneles de Power BI directamente en aplicaciones y sitios web personalizados. Esta opción de licenciamiento se basa en el consumo y está diseñada para escenarios en los que los usuarios finales interactúan con los informes incrustados, pero no requieren acceso completo a las capacidades de Power BI.
>* **Power BI Report Server**: Power BI Report Server es una solución de informes local de Microsoft que permite a las organizaciones alojar y administrar informes de Power BI en sus propios servidores locales. A diferencia de Power BI Service, que es la plataforma en la nube de Microsoft, Power BI Report Server es una opción para aquellos que prefieren mantener sus datos y contenido en sus propias infraestructuras.

## **<font color="#000000">¿Que es Power Query?</font>**

</td>
</tr>
</table>

<center>

imagen.png

</center>

Power Query es una herramienta de transformación y preparación de datos que forma parte de Power BI, una plataforma de análisis de datos y creación de informes desarrollada por Microsoft.

Power Query permite a los usuarios importar, transformar y combinar datos de diversas fuentes para crear conjuntos de datos limpios y listos para el análisis.

Con Power Query, los usuarios pueden realizar una serie de tareas de preparación de datos, como la limpieza de datos, la eliminación de filas duplicadas, la combinación de varias fuentes de datos, la creación de columnas personalizadas y la aplicación de transformaciones avanzadas.

La herramienta ofrece una interfaz gráfica intuitiva que permite a los usuarios realizar estas tareas sin necesidad de escribir código.

Power Query puede conectarse a una amplia gama de fuentes de datos, incluyendo bases de datos relacionales, archivos de Excel, archivos CSV, fuentes en línea y servicios en la nube como SharePoint, Salesforce y Azure.

Los datos importados a través de Power Query se cargan en el modelo de datos de Power BI, donde se pueden utilizar para crear informes interactivos y paneles de control.

## **<font color="#000000">Vistas de Power BI Desktop</font>**

</td>
</tr>
</table>

Las **Vistas en Power BI Desktop** están interconectadas y se pueden alternar entre ellas para realizar diferentes tareas durante el proceso de creación de informes y paneles de control.

Esto permite a los usuarios diseñar visualizaciones, modelar datos, crear medidas y transformar datos de manera eficiente y efectiva.

imagen.png

A continuación, se describen las principales vistas disponibles en Power BI Desktop:

>* **Vista de informe (Report view)**: Esta vista es la principal en Power BI Desktop y se utiliza para diseñar y crear informes interactivos. Permite agregar visualizaciones, como gráficos, tablas, mapas y otros elementos visuales, y configurar sus propiedades. En la vista de informe, se pueden aplicar filtros, crear relaciones entre diferentes conjuntos de datos y diseñar paneles de navegación.
>* **Vista de datos (Data view)**: En esta vista, se puede explorar y editar los datos subyacentes utilizados en el informe. Permite ver las tablas, columnas y filas de los conjuntos de datos importados o conectados. Además, se pueden realizar transformaciones de datos, como la limpieza, la agregación y la creación de columnas calculadas.
>* **Vista de modelo (Model view)**: Esta vista muestra el modelo de datos subyacente utilizado en el informe. Permite definir relaciones entre tablas, crear medidas calculadas utilizando el lenguaje DAX (Data Analysis Expressions) y administrar la estructura y las propiedades de los datos.
>* **Vista de relaciones (Relationship view)**: Esta vista proporciona una representación visual de las relaciones entre las tablas en el modelo de datos. Permite crear, editar y eliminar relaciones, y verificar la integridad y la dirección de las mismas.
>* **Vista de consulta (Query view)**: En esta vista, se puede acceder y editar el proceso de consulta utilizado para importar y transformar los datos. Permite realizar acciones como la limpieza de datos, la unión de tablas, la filtración y la aplicación de transformaciones avanzadas utilizando el Editor de consultas de Power Query.
